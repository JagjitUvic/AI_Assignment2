## Implementing the Double Dice Roll Function

```python
import numpy as np
import collections as coll
Jail = False
# function to implement double dice roll
def double_dice_roll(count,steps,execute):
    steps1 = steps
    # rolling for the number of times
    #print "Rolling Dice Number",count
    # decided on a cdf
    cdf = []
    # making the array of CDF
    for x in range(1, 7):
        cdf.append(x/float(6))
    # random dice numbers
    dice_1 = np.random.rand(1)
    dice_2 = np.random.rand(1)
    if (count <=3):
        # comparing them with cdf and printing dice values
        d1 = np.argmax(cdf>dice_1)+1
        d2 = np.argmax(cdf>dice_2)+1
        #print d1,d2
        steps1 = steps1 + d1+d2
        #print steps1
        if (d1 == d2):
            return (count+1,steps1,True)
        else:
            return (count+1,steps1,False)
    else:
        return (count+1,steps1,False)
# returns steps or jail
def dice_result():
    Jail = False
    steps = double_dice_roll(count = 1,steps = 0,execute=True)
    if (steps[2] == True):
        steps = double_dice_roll(steps[0],steps[1],steps[2]) 
    if (steps[2] == True):
        steps = double_dice_roll(steps[0],steps[1],steps[2]) 
    if (steps[2] == True):
        #print "3 Times a double, You go to JAIL" 
        Jail = True

    if (Jail == False):  
        #print "Number of turns : " , steps[0]-1
        #print "Number of Steps : " , steps[1]
        #print "Will you get another chance : " , steps[2]
        return (steps[1], False)
    else:
        #print "Number of turns : " , steps[0]-1
        #print "Number of Steps : " , 0
        #print "Will you get another chance : " , False
        return (steps[1], True)
```

## Chance Card Implementation

```python
# function to execute random chance card selection
def chance_selection():
    # all chance options
    options = ['Get out of Jail Free','Collect Money','Advance Somewhere','You Pay Money','You go to Jail']
    #ints the probabilities of each of these
    ints = [1,4,12,15,16]
    # decided on a cdf
    cdf = []
    # making the array of CDF
    for idx, val in enumerate(ints):
        cdf.append(val/float(16))
    # random chance
    chance_random = np.random.rand(1)
    chance = np.argmax(cdf>chance_random)
    print options[chance]
chance_selection()
```

```python
# function to execute random community chest card selection
def community_selection():
    # all community options
    options = ['Get out of Jail Free','Collect Money','Advance Somewhere','You Pay Money','You go to Jail']
    #ints the probabilities of each of these
    ints = [1,10,11,15,16]
    # decided on a cdf
    cdf = []
    # making the array of CDF
    for idx, val in enumerate(ints):
        cdf.append(val/float(16))
    # random chance
    community_random = np.random.rand(1)
    community = np.argmax(cdf>community_random)
    print options[community]
community_selection()
```
